# Funciones

Las [Funciones](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica) son los objetos fundamentales con los que tratamos en el [Clculo](https://es.wikipedia.org/wiki/C%C3%A1lculo_infinitesimal). Las mismas pueden ser representadas de diferentes maneras: por una [ecuacin](https://es.wikipedia.org/wiki/Ecuaci%C3%B3n), en una tabla, por un grfico, o en palabras. Se utilizan principalmente como modelos matemticos para representar fenmenos del mundo real.

La palabra *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* fue introducida en las [Matemticas](https://es.wikipedia.org/wiki/Matem%C3%A1tica) por [Leibniz](https://es.wikipedia.org/wiki/Gottfried_Leibniz), quien utilizaba este trmino para designar cierto tipo de frmulas matemticas. Una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)*  surge cada vez que una cantidad depende de otra. Ms precisamente la definicin de *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* es esencialmente la siguiente: Dados dos conjuntos de objetos, el conjunto X y el conjunto Y, una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* es una regla que asocia a cada objeto de X, uno y slo un, objeto en Y. El conjunto X se denomina el *dominio* de la *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)*. Los objetos de Y, asociados con los objetos en X forman otro conjunto denominado el *recorrido* de la *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)*. Generalmente se utilizan las letras $f, g, h, G$ y $H$ para designarlas. Si $f$ es una funcin dada y $x$ es un objeto de su dominio, la notacin $f(x)$ se
utiliza para designar el objeto que en el recorrido corresponde a $x$, en la *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* $f$, y se denomina el valor de la funcin $f$ en $x$. El smbolo $f(x)$ se lee, f de x. 

Muchas veces resulta til pensar en una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* como si fuera una mquina. Si $x$ est en el dominio de la funcin $f$, entonces cuando $x$ entra en la mquina, se acepta como una entrada y la mquina produce una salida $f(x)$ de acuerdo a la regla de la funcin. As, podemos pensar al *dominio* como el conjunto de todas las entradas posibles y al *recorrido* como el conjunto de todas las salidas posibles.

El mtodo ms comn para la visualizacin de una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* es su [grfica](https://es.wikipedia.org/wiki/Gr%C3%A1fica_de_una_funci%C3%B3n). Si $f$ es una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* con
dominio $D$, a continuacin, su [grfica](https://es.wikipedia.org/wiki/Gr%C3%A1fica_de_una_funci%C3%B3n) es el conjunto de [pares ordenados](https://es.wikipedia.org/wiki/Pares_ordenados). 

$$\{(x, f(x)) \mid x \in D \} $$

Aqu debemos tener en cuenta que el par $(x, f(x))$, es un par *entrada-salida*, el valor de $x$ representa el valor de entrada, mientras que el valor de $f(x)$ representa la salida de la *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)*.  En otras palabras, la [grfica](https://es.wikipedia.org/wiki/Gr%C3%A1fica_de_una_funci%C3%B3n) de $f$ se compone de todos puntos $(x, y)$ en el plano de coordenadas tal que $y=f(x)$ y $x$ est en el *dominio* de $f$.
La [grfica](https://es.wikipedia.org/wiki/Gr%C3%A1fica_de_una_funci%C3%B3n) de una *[Funcin](https://es.wikipedia.org/wiki/Funci%C3%B3n_matem%C3%A1tica)* $f$ nos da una imagen til del comportamiento o la "historia de vida" de la misma.

# Lmites

Uno de los conceptos ms importantes dentro del [Clculo](https://es.wikipedia.org/wiki/C%C3%A1lculo_infinitesimal) es el concepto de *[Lmite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)*. Se dice que una funcin $f$ tiende hacia el *[Lmite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)* $l$ cerca de $a$, si se puede hacer que $f(x)$ este tan prxima como queramos de $l$, haciendo que $x$ est suficientemente cerca de $a$, pero siendo distinta de $a$. As por ejemplo si analizamos la funcin $f(x) = x^2 - x + 2$, para los valores cercanos a 2, podramos ver los siguientes resultados.

```python
# <!-- collapse=True -->
def f(x):
    return x**2 - x + 2

x = np.array([1, 1.5, 1.9, 1.95, 1.99, 1.999, 2.001, 2.05, 2.1, 2.2, 2.5, 3 ])
y = f(x)
tabla = pd.DataFrame(list(zip(x, y)), columns=['x', 'f(x)'])
tabla
```

de acuerdo con esta tabla, podemos ver que a medida que hacemos al valor de $x$ cercano a 2, vemos que $f(x)$ se hace muy cercana a 4. Incluso podramos hacer a $f(x)$ tan cercana como queramos a 4, haciendo que $x$ este lo suficientemente cerca de 2. Por lo tanto, podemos expresar esta propiedad diciendo que el *"[Lmite](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico) de la funcin $f(x) = x^2 - x + 2$ cuando $x$ se acerca a 2 es igual a 4."* y lo podemos representar con la siguiente notacin:

$$\lim_{x\to 2} \left(x^2 -x + 2\right) = 4$$

Grficamente lo podemos ver del siguiente modo.

```python
# <!-- collapse=True -->
x = np.linspace(-2, 4, num=30)

ax = move_spines()
ax.grid()
ax.plot(x, f(x))
ax.scatter(2, 4, label="limite cuando x tiende a 2", color='r')
plt.legend()
plt.title(r"Grafico de $f(x)=x^2 -x + 2$")
plt.ylabel('f(x)')
plt.xlabel('x')
plt.show()
```

### Las leyes de los lmites

Calcular el valor exacto de los *[Lmites](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)* muchas veces no suele tan fcil como reemplazar el valor de $a$ en $f(x)$. Es por esto que es importante conocer algunas propiedades de los *[Lmites](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)*, ellas son:

1. **Ley de la suma:** El lmite de la suma de dos funciones es la suma de sus lmites.
2. **Ley de la diferencia:** El lmite de la diferencia de dos funciones es la diferencia de sus lmites.
3. **Ley del producto:** El lmite del producto de dos funciones es el producto de sus lmites.
4. **ley del mltiplo constante:** El lmite de una constante por una funcin es la constante por el lmite de la funcin.
5. **Ley del cociente:** El lmite del cociente de dos funciones es el cociente de sus lmites, siempre que el lmite del denominador sea diferente de *cero*.

Es decir que si tenemos a la constante $C$ y a los lmites $\lim_{x\to a} f(x)$ y $\lim_{x\to a} g(x)$. Entonces podemos expresar estas propiedades matemticamente de la siguiente forma:

1- **Ley de la suma:** $\lim_{x\to a} [f(x) + g(x)] = \lim_{x\to a} f(x) + \lim_{x\to a} g(x)$.


2- **Ley de la diferencia:** $\lim_{x\to a} [f(x) - g(x)] = \lim_{x\to a} f(x) - \lim_{x\to a} g(x)$.


3- **Ley del producto:** $\lim_{x\to a} [f(x) \cdot g(x)] = \lim_{x\to a} f(x) \cdot \lim_{x\to a} g(x)$.


4- **ley del multiplo constante:** $\lim_{x\to a} [C \cdot f(x)] = C \cdot \lim_{x\to a} f(x)$.


5- **Ley del cociente:** $\lim_{x\to a} \left[\frac{f(x)}{g(x)}\right] = \frac{\lim_{x\to a} f(x)}{\lim_{x\to a} g(x)}$, si $\lim_{x\to a} g(x) \ne 0$.


### Calculando Lmites con Python

Con [Python](https://www.python.org/), podemos resolver *[Lmites](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)* fcilmente utilizando la librera [SymPy](https://www.sympy.org/es/), la cual nos proporciona el objeto `Limit` para representarlos en [Python](https://www.python.org/). Su sintaxis es la siguiente: `Limit(funcin, variable, punto)`. Entonces para calcular el lmite de $f(x)$ cuando $x$ tiende a 0, debemos escribir:

`Limit(f(x), x, 0)`

Lo utilizamos de la siguiente forma:

```python
from sympy.interactive import printing
from sympy import Limit, limit, Symbol, S

# imprimir con notacin matemtica.
printing.init_printing(use_latex='mathjax') 

x = Symbol('x') # Creando el simbolo x.
Limit(x**2 - x + 2, x, 2) # Creando el objeto Limit
```

```python
# Resolviendo el Limite con el metodo doit()
Limit(x**2 - x + 2, x, 2).doit()
```

```python
# La funcion limit nos da directamente el resultado
limit(x**2 - x + 2, x, 2)
```

```python
# Resolviendo limite 1/x cuando x tiende a infinito
Limit(1/x, x, S.Infinity)
```

```python
Limit(1/x, x, S.Infinity).doit()
```

Como vemos, primero creamos el smbolo para representar a la variable x utilizando el objeto `Symbol`, y luego creamos nuestro lmite utilizando el objeto `Limit`. Por ltimo para resolver el lmite, simplemente llamamos al mtodo `doit()` sobre el objeto `Limit` que acabamos de crear. Tambin podemos calcular los *[Lmites](https://es.wikipedia.org/wiki/L%C3%ADmite_matem%C3%A1tico)* de valores de $x$ que tiendan hacia el [infinito](https://es.wikipedia.org/wiki/Infinito) utilizando la *clase* especial `S.Infinity` que nos proporciona [SymPy](https://www.sympy.org/es/).
